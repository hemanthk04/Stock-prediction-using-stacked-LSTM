```python
#Splitting the data set in to testing & training data sets
#As this is an LSTM model, every data before this impacts the next one.

train_size = int(len(df)*0.7)
test_size = len(df) - train_size
train_df = df[0:train_size,:]
test_df = df[train_size:len(df),:1]
```

```python
#as this type is depended upon the previous data,
#features of the output should be the previous one. feature steps is the number of independent variables for the dependent variables.
#ie. when if the feature steps number is 50, then 51st data is dependent on the previous 100 values. so we make a data set, 
#where x values are of the previous 50 ones
#and y value is the result.

import numpy as n
def create_DS(dataframe,feature_steps=1):
    dataX, dataY = [],[]
    for i in range(len(dataframe) - feature_steps-1):
        p = dataframe[i:(i+feature_steps),0]
        dataX.append(p)
        dataY.append(dataframe[i+feature_steps,0])
    return n.array(dataX),n.array(dataY)
```

```python
feature_steps = 100
Train_X, Train_Y = create_DS(train_df,feature_steps)
Test_X, Test_Y = create_DS(test_df,feature_steps)
```

```python
Train_X, Train_Y
```

```python
print(Train_X.shape,Train_Y.shape)
print(Test_X.shape,Test_Y.shape)
```

```python
Train_X = Train_X.reshape(Train_X.shape[0], Train_X.shape[1],1)
Test_X = Test_X.reshape(Test_X.shape[0], Test_X.shape[1],1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
# To start creating Stacked LSTM model
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences = True, input_shape = (100,1)))
model.add(LSTM(50, return_sequences = True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
model.summary()
```

```python
model.fit(Train_X,Train_Y,validation_data=(Test_X,Test_Y),epochs=100,batch_size=64,verbose=1)
```

```python
Train_Forecast = model.predict(Train_X)
Test_Forecast = model.predict(Test_X)
```

```python
#Scaling back to the original / so using Inverse scaling
Train_Forecast = scaler.inverse_transform(Train_Forecast)
Test_Forecast = scaler.inverse_transform(Test_Forecast)
Train_Forecast, Test_Forecast
```

```python
from sklearn.metrics import mean_squared_error
```

```python
math.sqrt(mean_squared_error(Train_Y, Train_Forecast))
```

```python
math.sqrt(mean_squared_error(Test_Y, Test_Forecast))
```
